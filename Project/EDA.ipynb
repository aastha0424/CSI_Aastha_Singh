##### b. Categorical Features

```python
import pandas as pd
import warnings
# Load the dataset (adjust path if needed)
df = pd.read_csv("student-por.csv")
# Suppress future warnings
warnings.filterwarnings("ignore", category=FutureWarning)
import seaborn as sns
import matplotlib.pyplot as plt

categorical_cols = df.select_dtypes(include='object').columns
n_cols = 3  # number of plots per row
n_rows = (len(categorical_cols) + n_cols - 1) // n_cols  # calculate rows needed

plt.figure(figsize=(n_cols * 5, n_rows * 4))  # adjust size as needed

for idx, col in enumerate(categorical_cols, 1):
    plt.subplot(n_rows, n_cols, idx)
    sns.countplot(x=col, data=df, palette="pastel")
    plt.xticks(rotation=45)
    plt.title(f"Distribution of {col}")
    plt.tight_layout()

plt.suptitle("üìä Categorical Feature Distributions", fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```

### 4Ô∏è‚É£ Outlier Detection

```python
for col in ['age', 'failures', 'absences', 'G1', 'G2', 'G3']:
    sns.boxplot(y=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

### üìå Insights on Outlier Detection
- **Absences**:
  - The `absences` feature is **right-skewed** with a large number of students having **0‚Äì10 absences**.
  - There are multiple **outliers with more than 30 absences**, which may indicate **chronic absenteeism** and could negatively impact performance.

- **Failures & G3 (Final Grade)**:
  - Both `failures` and `G3` show **significant clustering near 0**.
  - This indicates that a subset of students are **failing severely**, receiving the lowest possible scores.
  - These features may introduce **non-linear effects** or require **special handling** during modeling (e.g., separating failing students as a class).


### 5Ô∏è‚É£ Correlation Analysis

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

### üî• Correlation Heatmap Insights

The correlation matrix provides important insights into how various numerical features are related to each other and to the final grade (`G3`):

---

#### üéØ Strongly Correlated with `G3` (Final Grade)
- **G1 (First Period Grade)**: `0.83`
- **G2 (Second Period Grade)**: `0.92`

üìå These are **strong positive correlations**, indicating that previous academic performance is a **very strong predictor** of final grades.

---

#### üîª Negatively Correlated with `G3`
- **Failures**: `-0.39` ‚Üí Students with more past class failures tend to perform worse.
- **Absences**: `-0.13` ‚Üí Higher absenteeism is weakly correlated with lower performance.
- **Dalc (Workday alcohol use)**: `-0.18` ‚Üí Slight negative correlation; higher alcohol consumption may be detrimental.

---

#### ‚ûï Weak to Moderate Positive Correlations
- **Study Time**: `0.25` ‚Üí Slightly positive impact on grades.
- **Mother‚Äôs Education (Medu)**: `0.24`
- **Father‚Äôs Education (Fedu)**: `0.21`

These features suggest that **parental education and time spent studying** have a **moderate influence** on student success.

---

#### üßä Weak or No Correlation
- Features like `goout`, `health`, `freetime`, `famrel` show **little to no correlation** with `G3`.
- This suggests these lifestyle or social factors may not significantly impact academic outcomes directly.

---

### ‚úÖ Summary

- Prior grades (`G1`, `G2`) are the **strongest predictors** of `G3`.
- Behavioral and demographic features have **moderate to weak** influence.
- Features like `failures`, `absences`, and `Dalc` may be important for identifying at-risk students.

This analysis supports selecting `G1`, `G2`, `studytime`, `failures`, and `Medu/Fedu` as important features for regression modeling.


### 6Ô∏è‚É£ Relationship Exploration

##### a. Final Grade (G3) vs Studytime

```python
sns.boxplot(x='studytime', y='G3', data=df)
plt.title("Study Time vs Final Grade")
plt.show()
```

##### üìö Study Time vs Final Grade

- Students with **higher study time** tend to have **higher final grades**.
- **Low studytime (1)** group shows lower median and more failing grades.
- **Studytime = 3 or 4** groups perform better and more consistently.
- Overall, `studytime` is a useful feature for predicting student performance.


##### b. Parental Education vs Final Grade

```python
sns.boxplot(x='Medu', y='G3', data=df)
plt.title("Mother's Education vs Final Grade")
plt.show()

sns.boxplot(x='Fedu', y='G3', data=df)
plt.title("Father's Education vs Final Grade")
plt.show()
```

##### üßë‚Äçüéì Parental Education vs Final Grade

The boxplots above show how the final grade (`G3`) varies based on parents' education levels.

---

##### üìò Mother's Education (`Medu`)
- As the **education level of the mother increases**, the **median final grade also increases**.
- Students whose mothers have higher education (level 4) show:
  - Higher median performance
  - Slightly wider grade range, suggesting more variation but higher potential
- Students with no maternal education (`Medu = 0`) have lower and more consistent (narrow) grade outcomes.

---

##### üìó Father's Education (`Fedu`)
- A similar positive trend is observed:
  - **Higher `Fedu` is associated with better performance**.
  - Students with fathers at education level 4 have among the **highest median G3 scores**.
- The effect is **less pronounced** than mother's education but still noticeable.

---

##### ‚úÖ Conclusion
- **Parental education**, especially the **mother‚Äôs**, is positively associated with academic performance.
- These features are valuable for inclusion in predictive models.

#### üåê Internet Access vs Final Grade (G3)
